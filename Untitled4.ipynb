```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
x = range(6)
plt.plot(x, [y**3 + 2*y**2 +1 for y in x])
plt.show()
```

```python
x = range(1, 5)
plt.plot(x, [y**3 + 2*y**2 +1 for y in x])
plt.plot(x, [y*3.0 for y in x])
plt.plot(x, [y/3.0 for y in x])
plt.show()
```

```python
x = range(1, 5)
plt.plot(x, [y**3 + 2*y**2 +1 for y in x], x, [y*3.0 for y in x], x, [y/3.0 for y in x])
plt.show()
```

```python
x = np.arange(0, 2*3.14, 0.1)
y = np.sin(x)
plt.plot(x, y, label='sin(x)')
y = np.cos(x)
plt.plot(x, y, label='cos(x)')
plt.grid(True)
plt.title("Plot of Sine and Cosine Function")
plt.legend(loc="upper right")
plt.savefig('plots.png')
plt.show()
```
